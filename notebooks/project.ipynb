```python
import re

# Lista de archivos que deseas procesar
Filenames = [
    '../data/baby1990.html', 
    '../data/baby1992.html',
    '../data/baby1994.html',
    '../data/baby1996.html',
    '../data/baby1998.html',
    '../data/baby2000.html',
    '../data/baby2002.html',
    '../data/baby2004.html',
    '../data/baby2006.html',
    '../data/baby2008.html',

]
for filename in Filenames:
    # Leer el contenido del archivo HTML
    with open(filename, 'r', encoding='utf-8') as file:
        html_content = file.read()

    # 1. Encontrar el aÃ±o con una expresiÃ³n regular
    year_match = re.search(r"Popularity in (\d{4})", html_content)
    if year_match:
        year = year_match.group(1)
        print("AÃ±o:", year)
    print ("    ")

    # 2. Extraer nombres y clasificaciones con expresiones regulares
    names_data = re.findall(r"<tr align=\"right\"><td>(\d+)</td><td>(\w+)</td><td>(\w+)</td>", html_content)
    
    print(names_data)
    for rank, name_boy, name_girl in names_data:
        print (f"Ranking: {rank}, NiÃ±o: {name_boy}, NiÃ±a: {name_girl}")
    print(" ")

    # 3. Guardar nombres y clasificaciones en un diccionario
    names_dict = {}
    for rank, boy_name, girl_name in names_data:
        if  boy_name not in names_dict or int(rank) < int(names_dict[boy_name]) :
            names_dict[boy_name] = rank
        if girl_name not in names_dict or int(rank) < int(names_dict[girl_name]):
            names_dict[girl_name] = rank
        continue
    print (names_dict.items())
    for clave, valor in names_dict.items():
        print(valor, clave)

    print(" ")
    # 4. Crear una lista con los nombres ordenados alfabÃ©ticamente
    sorted_names = sorted(names_dict.keys()) # sorted = ordena los elementos pasados como parametro y te devuelve una lista
    print(sorted_names)
    for name in sorted_names:
        print (name)
    print ("ultimo punto")
    # 5. Crear una lista con el formato especificado
    formatted_list = [f"AÃ±o: {year}"]

    formatted_list_name= []

    for rank, boy_name, girl_name in names_data:
        formatted_list_name.append(f"{boy_name} PosiciÃ³n:{rank}")
        formatted_list_name.append(f"{girl_name} PosiciÃ³n:{rank}") #.append = agregar

    formatted_list_name.sort() # .sort = ordenar una lista
    formatted_list.extend(formatted_list_name) # .extend = concatenar

    # Imprimir el resultado
    for item in formatted_list:
        print(item)
```
